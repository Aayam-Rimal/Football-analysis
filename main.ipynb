```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("fifa_players.csv")
df=pd.DataFrame(data)

df.head(5)
```

```python
df.describe()
```

```python
df = df.drop(columns=['national_team','national_rating','national_team_position','national_jersey_number'], errors='ignore')
```

```python
df=df.fillna(0)
df.columns
```

```python
dribbler = df.loc[df['dribbling'].idxmax(), 'name']
print(f"best dribbler: {dribbler}")

penalties= df.loc[df['penalties'].idxmax(), 'name']
print(f"most penalties: {penalties}")

vision= penalties= df.loc[df['vision'].idxmax(), 'name']
print(f"best vision: {vision}")

overall= df.loc[df['overall_rating'].idxmax(), 'name']
print(f"best rating:{overall}")

finish= df.loc[df['finishing'].idxmax(), 'name']
print(f"best finishing: {finish}")

least_expensive= df.loc[df['value_euro'].idxmin(), 'name']
print(f"least value: {least_expensive}")



```

```python
rating= df.groupby('nationality')['overall_rating'].mean()
maxrating= rating.idxmax()
minrating= rating.idxmin()
print(f"max average rating: {maxrating}")
print(f"min average rating: {minrating}")
rating.sort_values(ascending=False)

```

```python
pivot= pd.pivot_table(df, index='name', values=['overall_rating','dribbling','vision','wage_euro'], aggfunc='max')
pivot.sort_values('wage_euro', ascending=False)
```

```python
print(f"average weight of  footballers: {df['weight_kgs'].mean()}")
print(f'average height of footballers: {df['height_cm'].mean()}')
print(f'average wage of footballers: {df['wage_euro'].mean()}')
print(f'average rating of footballers:{df['overall_rating'].mean()}')
```
